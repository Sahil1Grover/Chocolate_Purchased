```python
def price(data):
  if data['Brand']==1:
    return data['Quantity']*data['Price_1']
  elif data['Brand']==2:
    return data['Quantity']*data['Price_2']
  elif data['Brand']==3:
    return data['Quantity']*data['Price_3']
  elif data['Brand']==4:
    return data['Quantity']*data['Price_4']
  elif data['Brand']==5:
    return data['Quantity']*data['Price_5']
  else:
    return 0
data=df
df['Final_Price']=df.apply(price,axis=1)
```

```python
# NOW LETS CHECK WHETHER PROMOTION HELPED GENERATE ANY REVENUE OR NOT
# LETS CREATE A PRICE COLUMN WITHOUT PROMOTION AND A COLUMN WITH PROMOTION

def price(data):
  if data['Brand']==1 and data['Promotion_1']==0:
    return data['Quantity']*data['Price_1']
  elif data['Brand']==2 and data['Promotion_2']==0:
    return data['Quantity']*data['Price_2']
  elif data['Brand']==3 and data['Promotion_3']==0:
    return data['Quantity']*data['Price_3']
  elif data['Brand']==4 and data['Promotion_4']==0:
    return data['Quantity']*data['Price_4']
  elif data['Brand']==5 and data['Promotion_5']==0:
    return data['Quantity']*data['Price_5']
  else:
    return 0
data=df
df['Price_Without_Prom']=df.apply(price,axis=1)
```

```python
without_promo=df.groupby('Brand')['Price_Without_Prom'].sum().reset_index(name='Price_Without_Prom')
without_promo
```

```python
def price(data):
  if data['Brand']==1 and data['Promotion_1']==1:
    return data['Quantity']*data['Price_1']
  elif data['Brand']==2 and data['Promotion_2']==1:
    return data['Quantity']*data['Price_2']
  elif data['Brand']==3 and data['Promotion_3']==1:
    return data['Quantity']*data['Price_3']
  elif data['Brand']==4 and data['Promotion_4']==1:
    return data['Quantity']*data['Price_4']
  elif data['Brand']==5 and data['Promotion_5']==1:
    return data['Quantity']*data['Price_5']
  else:
    return 0
data=df
df['Price_With_Prom']=df.apply(price,axis=1)
```

```python
with_promo=df.groupby('Brand')['Price_With_Prom'].sum().reset_index(name='Price_With_Prom')
with_promo
```

```python
# let us merge this dataset
prom_analysis=pd.merge(with_promo,without_promo, on='Brand')
prom_analysis
```

```python
#Plotting the data side by side on the graph
ax=prom_analysis.plot(x="Brand", y=["Price_With_Prom","Price_Without_Prom"], kind="bar")
plt.title('Promotion_Analysis')
plt.xlabel('Brand')
plt.ylabel('Revenue')
for i in ax.containers:
  ax.bar_label(i,)

plt.show()
```

From this grapical representation we can infer that applying promotion on chocolate brands did not help in increasing revenue. The revenue fell drastically instead.
